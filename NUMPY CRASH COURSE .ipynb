 # Numpy masking or filter

```python
mat = np.arange(0,100).reshape(10,10)
mat
```

```python
id(mat)
```

```python
mat.shape
```

```python
mat < 50    # it prints true and false
```

```python
mat[mat < 50]      # its print the values which are less than 50
```

```python
mat[mat <= 50]
```

```python
mat[mat !=50]     # its print all the values except 50
```

```python
mat[mat > 50]      # it prints the value which are greater than 50
```
